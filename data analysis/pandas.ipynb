## Boolean Indexing
`df['column'] > value`

`df['column'] == value`

`df['column'] != value`

```python
df[df['price'] > 300000]
```

```python
df[df['condition'] >= 5]
```

```python
# 5 star rated house + 2 floors
df[(df['condition'] == 5) & (df['floors'] == 2)]
```

```python
f1 = df['price'] > 100000
f2 = df['condition'] == 5
f3 = df['floors'] == 2
df[f1 & f2 & f3]
```

```python
df[f1 & f2 & f3].reset_index().style.highlight_max(color='blue')
```

```python
df[f1 & f2 & f3].reset_index().style.background_gradient(cmap='cool')
```

```python
df[f1 & f2 & f3].reset_index().style.set_properties(color='red', subset=['price'])
```

```python
# query
df.query('price > 250000')
```

```python
df.query('price > 2500000 & condition == 5')
```

```python
df.query('price > 2500000 & condition == 5').sort_values('price')
```
