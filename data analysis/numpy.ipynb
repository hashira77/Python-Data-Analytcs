# 2d Slicing Matrix

```python
import numpy as np
```

```python
x = np.random.randint(1,10,(5,5)) # 5*5 matrix
print(x)
```

```python
# row first (index 0)
print(x[0])
print(x[0,-1]) # row 1st col 2nd
print(x[-1]) # last row
print(x[:3]) # first 3 rows
print(x[3:]) # all roes after 3rd row
```

```python
print(x[:,0]) # : all rows, 0 column index
```

```python
print(x[:,-1])
```

```python
print(x[:,:2])
```

```python
print(x[:,[0,1,-1]]) # selective column index
```

```python
print(x[1:-1,1:-1]) # middle order
```

# Boolean Filtering

```python
x[x > 5] # Boolean Filtering/Indexing
```

```python
y = x [x <= 6]
print(y.shape)
z= y.reshape(6,4)
print(z)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plt.figure(figsize=(10,10))
plt.plot([1,2,3], [4,5,6], linewidth = 20, color = 'orange', linestyle = 'dashdot', marker = 'o', markersize = 20)
plt.plot([1,2,3], [6,5,4], linewidth = 20, color = 'blue', linestyle = 'dashdot', marker = 'o', markersize = 20)
plt.xticks([1,2,3], ['one', 'two', 'three'], rotation = 45)
plt.yticks([4,5,6], ['four', 'five', 'six'], rotation = 90)
plt.title('My First Plot')
plt.xlabel('X axis')
plt.ylabel('Y-axis')
plt.text(x = 2, y = 5.1, s = "Crossing Point", fontsize = 12, color = 'red', horizontalalignment = 'center')
plt.show()
```

Standard Way of Visualization

```python
fig, ax = plt.subplots (nrows = 1, ncols = 2, figsize = (10,5))
t = np.arange(-10,10,.1)
sine = np.sin(t)
cosine = np.cos(t)
# print(t)
# print(sine)
# print(cosine)
ax[0].plot(t, sine, label = 'sine', color = 'red', linestyle = '', marker = 'x')
ax[1].plot(t, cosine, label = 'cosine', color = 'green')
plt.savefig('sine_cosine.png', dpi = 300)
```
