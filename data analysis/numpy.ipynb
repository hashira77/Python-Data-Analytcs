# 2d Slicing Matrix

```python
import numpy as np
```

```python
x = np.random.randint(1,10,(5,5)) # 5*5 matrix
print(x)
```

```python
# row first (index 0)
print(x[0])
print(x[0,-1]) # row 1st col 2nd
print(x[-1]) # last row
print(x[:3]) # first 3 rows
print(x[3:]) # all roes after 3rd row
```

```python
print(x[:,0]) # : all rows, 0 column index
```

```python
print(x[:,-1])
```

```python
print(x[:,:2])
```

```python
print(x[:,[0,1,-1]]) # selective column index
```

```python
print(x[1:-1,1:-1]) # middle order
```

# Boolean Filtering
- Extract data from matrix based on condition

```python
x > 5 # Boolean Matrix
```

```python
x[x > 5] # Boolean Filtering/Indexing
```

```python
y = x [x <= 6]
print(y.shape)
z= y.reshape(6,4)
print(z)
```
